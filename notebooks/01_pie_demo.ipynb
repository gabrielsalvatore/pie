```python
import sys
from pathlib import Path

here = Path().resolve()
print("Notebook dir:", here)

# Walk upward until we find a folder that contains "src"
root = None
for p in [here] + list(here.parents):
    if (p / "src").exists():
        root = p
        break

print("Detected project root:", root)

if root is None:
    raise RuntimeError("Could not find a parent directory containing 'src'.")

if str(root) not in sys.path:
    sys.path.insert(0, str(root))

print("Added to sys.path:", root)
```
